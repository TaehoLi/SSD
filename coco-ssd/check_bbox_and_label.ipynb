## check bbox and label in COCO

```python
import json
from collections import OrderedDict
import pathlib
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


root = "/home/taeho/data/coco2014"
root = pathlib.Path(root)
annotation_file = root / f"annotations/annotations/instances_val2014.json"

def check_bbox_or_label(image_id):
    with open(annotation_file, encoding="utf-8") as data_file:    
        data = json.load(data_file, object_pairs_hook=OrderedDict)

    boxes = []
    labels = []
    is_difficult = []
    
    
    #image_id = "COCO_val2014_000000005802"
    image_number = int(image_id[-6:])
    print(image_number)
    
    # get image_height and width
    for index, name in enumerate(data["images"]):
        if name["id"] == image_number:
            image_height = name["height"]
            image_width = name["width"]

    for index, name in enumerate(data["annotations"]):
        category_id = name["category_id"]
        bbox = name["bbox"]

        if name["image_id"] == image_number:

            for index, name in enumerate(data["categories"]):
                if name["id"] == category_id:
                    xmin = round(bbox[0])
                    if xmin == 0:
                        xmin += 1
                    ymin = round(bbox[1])
                    if ymin == 0:
                        ymin += 1
                    xmax = round(bbox[0] + bbox[2])
                    if xmax == image_width:
                        xmax -= 1
                    ymax = round(bbox[1] + bbox[3])
                    if ymax == image_height:
                        ymax -= 1
                        
                    bbox = [xmin, ymin, xmax, ymax]
                    boxes.append(bbox)
                    
                    if category_id <= 11:
                        labels.append(category_id)
                    elif category_id <= 25:
                        labels.append(category_id-1)
                    elif category_id <= 28:
                        labels.append(category_id-2)
                    elif category_id <= 44:
                        labels.append(category_id-4)
                    elif category_id <= 65:
                        labels.append(category_id-5)
                    elif category_id == 67:
                        labels.append(61)
                    elif category_id <= 70:
                        labels.append(62)
                    elif category_id <= 82:
                        labels.append(category_id-9)
                    else:
                        labels.append(category_id-10)

    return (np.array(boxes, dtype=np.float32),
            np.array(labels, dtype=np.int64))

print(check_bbox_or_label("COCO_val2014_000000223648"))
print(check_bbox_or_label("COCO_val2014_000000204805"))
"""
print(check_bbox_or_label("COCO_val2014_000000113588"))
print(check_bbox_or_label("COCO_val2014_000000384553"))
print(check_bbox_or_label("COCO_val2014_000000337264"))
print(check_bbox_or_label("COCO_val2014_000000368402"))
print(check_bbox_or_label("COCO_val2014_000000012448"))
print(check_bbox_or_label("COCO_val2014_000000252219"))
print(check_bbox_or_label("COCO_val2014_000000079841"))
print(check_bbox_or_label("COCO_val2014_000000087038"))
print(check_bbox_or_label("COCO_val2014_000000174482"))
print(check_bbox_or_label("COCO_val2014_000000515289"))
print(check_bbox_or_label("COCO_val2014_000000562150"))
print(check_bbox_or_label("COCO_val2014_000000542145"))
print(check_bbox_or_label("COCO_val2014_000000403385"))
print(check_bbox_or_label("COCO_val2014_000000412151"))
print(check_bbox_or_label("COCO_val2014_000000540186"))
print(check_bbox_or_label("COCO_val2014_000000579003"))
print(check_bbox_or_label("COCO_val2014_000000242611"))
print(check_bbox_or_label("COCO_val2014_000000269105"))
print(check_bbox_or_label("COCO_val2014_000000051191"))
print(check_bbox_or_label("COCO_val2014_000000462565"))
"""
print(check_bbox_or_label("COCO_val2014_000000060760"))
print(check_bbox_or_label("COCO_val2014_000000294832"))
```
