```python
import xml.etree.ElementTree as ET
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

path = "/home/taeho//NEU-DET/ANNOTATIONS/"

xmin = []
ymin = []
xmax = []
ymax = []

image_path = "data/patches_294.jpg"

a = image_path.split('/')
b = a[1].split('.')
xml = b[0] + ".xml"
print(xml)

tree = ET.parse(path + xml)
root = tree.getroot()

for child in root.iter():
    tag = child.tag
    if tag == "xmin":
        xmin.append(int(child.text))
    if tag == "ymin":
        ymin.append(int(child.text))
    if tag == "xmax":
        xmax.append(int(child.text))
    if tag == "ymax":
        ymax.append(int(child.text))

print(xmin)
print(ymin)
print(xmax)
print(ymax)

orig_image = cv2.imread("./data/patches_294.jpg")
image = cv2.cvtColor(orig_image, cv2.COLOR_BGR2RGB)

for i, num in enumerate(xmin):
    cv2.rectangle(image, (xmin[i], ymin[i]), (xmax[i], ymax[i]), (0, 255, 0), 4)

#cv2.imshow("test", image)
#cv2.waitKey()
#cv2.destroyAllWindows()

plt.figure(figsize=(4,4))
plt.axis('off')
plt.xticks([]), plt.yticks([])
plt.tight_layout()
plt.imshow(image)
plt.show()
```
